```python
#lowess example
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.api import nonparametric
```

```python
x = np.linspace(-10,10)
y = x**2
```

```python
# add random error on x
a = 0.5
b = 1.0
data_x = []
for i,xi in enumerate(x):
    coin = np.random.randint(0,2)
    elem = xi+(-1)**(coin)*((b-a)*np.random.random_sample() + a)
    data_x.append(elem)
```

```python
# parabola
data_y = np.array(data_x)**2
```

```python
plt.plot(data_y, 'o', label= 'data with error to fit')
plt.plot(y, color = 'grey', ls = 'dashed', label = 'unknown relation to test fit')
plt.legend()
```

```python
# lowess fit
lowess_data_y = nonparametric.lowess(data_y,range(len(data_y)))
```

```python
fig = plt.figure()
plt.plot(data_y, 'o', label= 'data with error to fit')
plt.plot(y, color = 'grey', ls = 'dashed', label = 'unknown relation to test fit')
plt.plot(lowess_data_y[:,0], lowess_data_y[:,1], color = 'red', label = 'lowess')
plt.legend(loc = 'best');
```
